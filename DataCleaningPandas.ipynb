# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\danie\OneDrive\Escritorio\RoadMap Data Analys\Python\Datas\Customer Call List.xlsx")
df
```

```python
# eliminar duplicados
df = df.drop_duplicates()
df
```

```python
# eliminar columnas
df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
# strip en Last_Name (eliminar caracteres que no queremos)
df['Last_Name'] = df['Last_Name'].str.strip('123._/')
df
```

```python
# limpiar los numeros
# forma 1
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
#df

# forma 2 pero primero hay que cambiar la columna de phone_number a tipo numerico
#df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
#df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
#df

df['Phone_Number'] = df['Phone_Number'].fillna('')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na','', regex=True)
df
```

```python
df
```

```python
# separar columnas
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(",", n=2, expand = True)
df
```

```python
# estandarizar los valores de columnas usando replace
df['Paying Customer'] = df['Paying Customer'].replace({'Y':'Yes','N':'No'})

df['Do_Not_Contact'] = df['Do_Not_Contact'].replace({'Y':'Yes','N':'No'})
df
```

```python
# eliminar vacios
df = df.fillna()
#df = df.replace('N/a','')
```

```python
df.set_index('CustomerID', inplace=True)
```

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Yes':
        df.drop(x, inplace=True)
        
df
```

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace=True)
        
df

# otra forma de borrar nulos
df = df.dropna((subset='Phone_Number'), inplace=True)
```

```python
df = df.reset_index(drop=True)
df
```
