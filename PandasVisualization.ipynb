# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\danie\OneDrive\Escritorio\RoadMap Data Analys\Python\Datas\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('ggplot')
```

## Line Plot

```python
df.plot(kind = 'line', title = 'Ice Cream Rating', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

## Bar Plot

```python
df.plot.barh(stacked=True)
```

## Scatter Plot

```python
df.plot.scatter(x='Texture Rating', y='Overall Rating', s=100, c='Green')
```

## Histograma Plot

```python
df.plot.hist(bins=10)
```

## Box Plot

```python
df.boxplot()
```

## Area Plot

```python
df.plot.area(figsize=(5,5))
```

## Pie Chart

```python
df.plot.pie(y='Flavor Rating', figsize=(10,5))
```
